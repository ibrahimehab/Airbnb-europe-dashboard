# Bronze Layer

```python
import pandas as pd
from pathlib import Path

RAW_DIR = Path("../0th raw_source_files")
BRONZE_DIR = Path("raw")
BRONZE_DIR.mkdir(parents=True, exist_ok=True)
```

```python
files = list(RAW_DIR.glob("*.csv"))
print(f"Found {len(files)} CSV files.")
files
```

```python
for file in files:
    name = file.stem
    city, day_type = name.rsplit("_", 1)

    df = pd.read_csv(file)
    df["city"] = city.capitalize()
    df["day_type"] = day_type

    df.to_csv(BRONZE_DIR / f"{name}.csv", index=False)
    print(f"  -> {name}: {df.shape[0]} rows ingested")

print("Bronze ingestion complete.")
```
